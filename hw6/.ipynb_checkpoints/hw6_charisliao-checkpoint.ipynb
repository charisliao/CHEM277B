We will now examine unsupervised learning for classification for the data set of chemical compounds. **(b) Do KMeans clustering with K=2,3 and 4 clusters. Visualize your result (you can select 2 features to do visualization) and comment on which K value make the most sense to you according to the visualization you see. (Use the provided code if you are a ugrad or fill in the code for KMeans if you are a grad.)**

```python
import warnings

class KMeans():
    def __init__(self, K, maximum_iters=100):
        # K: number of clusters to be created
        # distance matrix is Eucledian distance
        self.K = K
        self.maximum_iters = maximum_iters
    
    def cluster(self, input_points):
        """ Do KMeans clustering
        input_points: np.array shape(ndata,nfeatures). 
            Each feature is assumed to be normalized within range of [0,1]
        """
        centroids = np.random.random((self.K, input_points.shape[1]))
        assignments = np.zeros_like(input_points.shape[0])
        new_assignments = self.create_new_assignments(centroids, input_points)
        
        # restart if run into bad initialization
        # Comment out this part for Q1.(d)
        if len(np.unique(new_assignments))<self.K:
            return self.cluster(input_points)
        
        n_iters = 1
        while (new_assignments != assignments).any() and n_iters < self.maximum_iters:
            ### Compute the centroid given new assignment ###
            centroids = np.array([np.mean(input_points[new_assignments==i], axis=0) for i in range(self.K)])
            assignments = new_assignments
            
            ### Update the assignment with current centroids ###
            new_assignments = self.create_new_assignments(centroids, input_points)
            if len(np.unique(new_assignments))< self.K:
                warnings.warn('At least one centroid vanishes')
            n_iters += 1
            if n_iters == self.maximum_iters:
                print("Warning: Maximum number of iterations reached!")
                
        return new_assignments
            
        
    def create_new_assignments(self, centroids, data_points):
        """ Assign each datapoint to its nearest centroid.
        centroid: 2d array of the current centroid for each cluster
        data_points: 2d arrays recording the features of each data point.
        """
        ###Compute the distances that stores the Eucledian distances between each datapoints and the centroid ###
        #shape (ndata,ncentroid)
        distances = np.array([np.linalg.norm(data_points - centroid, axis=1) for centroid in centroids]).T
#         print(distances.shape)
        new_assignments = np.argmin(distances, axis=-1)
        return new_assignments
        
   
        
```

```python
compound_features = scaled_features.drop(['type'], axis = 1)

# print(compound_features.shape)

# Create an instance of the KMeans class with K=2, 3, 4
kmeans2 = KMeans(2)
kmeans3 = KMeans(3)
kmeans4 = KMeans(4)

# Perform KMeans clustering with each K value
clusters2 = kmeans2.cluster(compound_features.values)
clusters3 = kmeans3.cluster(compound_features.values)
clusters4 = kmeans4.cluster(compound_features.values)

# plot data with different cluster assignments
sns.scatterplot(x='A', y='C', hue=clusters2, data=compounds)
plt.title('KMeans Clustering with K=2')
plt.show()

sns.scatterplot(x='A', y='C', hue=clusters3, data=compounds)
plt.title('KMeans Clustering with K=3')
plt.show()

sns.scatterplot(x='A', y='C', hue=clusters4, data=compounds)
plt.title('KMeans Clustering with K=4')
plt.show()
```

From the scatterplots above, we can see the with K=2 clusters, the data is split into two groups. When K=3 clusters, data is split into three groups. when K=4 clusters, data is split into four groups. Based on the observations, I think K=3 clusters makes the most sense because the clusters are clear and easy to identify but not too simple with just only two clusters. 

**(c) For K=3 clustering result, compare it to the true data label. How good is the classification?** 

```python
def validate(y_hat,y):
    """print accuracy of prediction for each class for the compounds dataset
    yhat: np.array shape(ndata). Your prediction of classes
    y: np.array of str shape(ndata). data labels / groudn truths.
    """
    # correct classification
    compounds = np.unique(y) # should be ['amide','phenol','ether'] for compounds dataset
    clusters =[np.where((y==c)) for c in compounds]
    pred_class = np.unique(y_hat)
    
    #remove -1 for noise point in DBSCAN
    pred_class= np.delete(pred_class,np.where(pred_class==-1))
    assert len(pred_class) == len(compounds), f'y_hat has less or more than {len(compounds)} classes:{pred_class}'

    for i in range(3):
        #loop over solutions
        counts=[]
        scores=[]
        for j in range(3):
            #loop over clusters of true assignments
            sol_i= np.where((y_hat==pred_class[i]))
            counts.append(len(np.intersect1d(sol_i, clusters[j])))
            scores.append(counts[-1]/len(clusters[j]))
        idx = np.argmax(scores)
        print(f'Class {pred_class[i]} - {compounds[idx]}: {counts[idx]} out \
of {np.count_nonzero(clusters[idx])} are classified correctly')
```

```python
validate(clusters3, compounds['type'])
```

```python
sns.scatterplot(x='A', y='C', hue=clusters3, data=compounds)
plt.title('KMeans Clustering with K=3')
plt.show()

sns.relplot(data=scaled_features, x="A", y="C",hue=scaled_features['type'])
plt.title('True data label')
plt.show()
```

The classification performs quite well with phenol being classified 100% accurately, amide being classified 97.09% correctly, and ether being classified 92%. From comparing the two plots above, we can see that the classification performs quite well as well.
